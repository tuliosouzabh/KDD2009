### Realizando operações com as colunas do dataframe

```python
## Para isto vamos criar um dataframe
data = {'indice': ['Obs 1', 'Obs 2', 'Obs 3'],
       'v1': [10,20,30],
       'v2': [50,60,70]}

bd = pd.DataFrame(data, columns= ['indice', 'v1', 'v2'])
bd.head()
```

```python
#Adição
bd['soma'] = bd['v1'] + bd['v2']
bd['soma+10'] = bd['soma'] + 10
bd.head()
```

```python
#usando o pandas add
bd['soma'] = bd['v1'].add(bd['v2'])
bd['soma+10'] = bd['soma'].add(10)
bd.head()
```

```python
#Subtração
bd['subtracao'] = bd['v2'] - bd['v1']
bd['subtracao10'] = bd['subtracao'] - 10
bd.head()
```

```python
#Usando o pandas sub
bd['subtracao'] = bd['v2'].sub(bd['v1'])
bd['subtracao10'] = bd['subtracao'].sub(10)
bd.head()
```

```python
#Multiplcação
bd['mult'] = bd['v2']*bd['v1']
bd['mult10'] = bd['mult']*10
bd.head()
```

```python
#Usando o pandas mul
bd['mult'] = bd['v2'].mul(bd['v1'])
bd['mult10'] = bd['mult'].mul(10)
bd.head()
```

```python
# Divisao
bd['div'] = bd['v1'] / bd['v2']
bd['div10'] = bd['div'] / 10
bd.head()
```

```python
#Usando o pandas div
bd['div'] = bd['v1'].div(bd['v2'])
bd['div10'] = bd['div'].div(10)
bd.head()
```

```python
#Somando todos os valores das colunas do dataframe
bd.sum()
```

```python
#Somando todos os valores de uma coluna especifica
coluna = 'v1'
bd[coluna].sum()
```

```python
#Maior valor das colunas
bd.max()
```

```python
#Maior valor de uma coluna especifica
coluna = 'v1'
bd[coluna].max()
```

```python
#Minimo valor do dataframe
bd.min()
```

```python
#Menor valor de uma coluna especifica
coluna = 'v1'
bd[coluna].min()
```

## Imputação de dados faltantes

```python
#Vamos calcular a média e imputar nos dados faltantes da variável Idade
#Somando todas as colunas do banco de dados
df['Age'].sum()
```

```python
#Contando os valores validos da variavel Idade
df['Age'].count()
```

```python
#Criando uma variavel com a média de idade
df['Media'] = df['Age'].sum() / df['Age'].count()
print(df['Media'][0])
```

```python
#Usando o comando fillna para preencher os dados faltantes com a média
df['Age'] = df['Age'].fillna(df['Media'])
```

```python
#Contando os valores validos da variavel Idade
df['Age'].isna().value_counts()
```
